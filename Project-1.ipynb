```python
"""
Purpose: Search and record words in a text.
Author: Long Pham, Phuc Nguyen.
Date: January 26, 2023 .
"""

import string

def format(text):
    """Normalization (lowercase characters and remove punctuations) the given text.

    :param: text - the inputted text to be normalize.
    :return: normalized text.
    :ptype: str
    :rtype: str
    """
    punc = string.punctuation
    for word in text:
        if word in punc:
            text = text.replace(word, "")
    text = text.lower()
    return text

def word(text):
    """Print out all words appear in the text.

    :param: text - the text to be processed.
    :return: list of words in the text.
    :ptype: str
    :rtype: list
    """
    text = format(text)
    text = text.split()
    text.sort()
    word_list = []
    for word in text:
        if word not in word_list:
            word_list.append(word)
    for word in word_list:
        print(word)
    return word_list

def findSentenceWithWord(text):
    """Prompt user to input a word, print out the sentences that have the word. If the word is not found in a text, notify the user that the word is not found.
    
    :param: text - the text to be processed.
    :return: none
    :ptype: str
    """
    
    sentencesList = text.split('.')
    sentencesList.pop()
    wordList = format(text)
    wordList = wordList.split()
    
    prompt = input("Enter a word: ").lower()
    
    if prompt not in wordList:
        print("Sorry, word not found.") 
    else:
        for i in range(len(sentencesList)):
            sentencesList[i] = sentencesList[i].strip()
            words = sentencesList[i].split(' ')

            for j in range(len(words)):
                if prompt == words[j].lower():
                    print(str(i + 1) + ' ' + sentencesList[i])
                    break
                
def main():
    """Access the input file and process it.

    :param: none
    :return: none
    """
    file = open("input.txt", "r",encoding = "utf-8")
    text = file.read()
    word(text)
    findSentenceWithWord(text)

if __name__ == "__main__":
    main()
```
